```python
drop_failures = False # to remove data from failure or recovery periods from training

#training data params
use_train_cache = True
maxv = 136500
step = 10500
window_length = 1050
stride = 10
traindir_loc = '/home/coder/processed/multivar/'

#model caching
use_cached_model = False
model_path = '/home/coder/model/multivar'
```

```python
pump_csv = pd.read_csv('~/pump_sensor_dataset/sensor.csv') \
  .drop(['Unnamed: 0','sensor_15','sensor_50'], axis=1)

df_pump  = pump_csv \
  .set_index(pd.to_datetime(pump_csv['timestamp'])) \
  .drop('timestamp', axis=1) \
  .asfreq('1T')

df_machine_status = pd.get_dummies(df_pump['machine_status']).reset_index(drop=True)

df_pump.drop('machine_status', axis=1, inplace=True)
```

```python
df_pump.fillna(0, inplace=True)
```

```python
def slide_window(df, window_length, stride=1, verbose = 1):
    orig_TS_list = []
    X_list = []
    series = df.copy()
    for i in series.columns.values: # loop through all input dimensions
        s = series[i]
        s2 = roll_fast(s.values, window_length)
        X_list.append(s2)
    X = numpy.dstack((X_list))
    if stride > 1:
        # If the training sequences are long, it does not make sense to move the sliding window with a stride of 1
        # since the training would take forever and the results would probably not be much better...
        X = X[::stride]
    if verbose > 2:
        print("X.shape:", X.shape)
    return X
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df_pump)
df_scaled = pd.DataFrame(scaled, columns=df_pump.columns.values)
df_scaled.shape
```

```python
df_scaled = pd.concat([df_scaled, df_machine_status], axis=1)
```

```python
m = (df_scaled['BROKEN']==1) | (df_scaled['RECOVERING']==1)
df_scaled[m][['BROKEN','RECOVERING']].plot(figsize=(15,5))
plt.axvline(x=136500, color='red', linestyle='--')
```

Train on 4 failures and test on the last 2: training on the first 136500 samples and test on the remaining. Zoomed in shows that 136500 is just after that 4th failure and recovery.

```python
df_train = df_scaled[:136500]
df_test = df_scaled[136500:]
df_train.shape, df_test.shape
```

```python
if drop_failures:
    #drop failure and recovery periods from training
    m = df_train['NORMAL'] == 1
    df_train = df_train[m]
df_train.shape
```

```python
# add batch dimension to test data so model can process it. 
test_X = df_test.values.reshape((1, df_test.shape[0], df_test.shape[1]))
df_test.shape, test_X.shape
```

```python
pd.DataFrame(test_X[0]).plot(figsize=(15,3), legend=False, title="TEST DATA")
```

```python
# create batched files for training - need to batch this due to memory constraint
if not use_train_cache:
    for start in numpy.arange(0,maxv,step):
        if start > len(df_train):
            break
        end = start + step
        if end > len(df_train):
            break
        batch = slide_window(df_train[start:end], window_length=window_length, stride=stride)
        filepath = f'{traindir_loc}{str(start).zfill(8)}-{str(end).zfill(8)}.npy'
        numpy.save(filepath, batch)
        print(f'{filepath}')
```

```python
file_list = tuple(sorted(os.listdir(traindir_loc)))
```

```python
filepathlist = tuple([y + x for x, y in zip(file_list, repeat(traindir_loc, len(file_list)))])
```

```python
train_X = numpy.stack(tuple([numpy.load(x) for x in filepathlist])) \
    .reshape(len(filepathlist)*numpy.load(filepathlist[0]).shape[0], window_length, numpy.load(filepathlist[0]).shape[2])
```
